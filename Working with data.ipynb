## Working with CSV

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import sys
# If we omitt the header, the first row would be consider the header, default for sep is ','
# We can pass the 'nrows' parameter with the number of rows to read if the file is very big
dframe = pd.read_csv('lec25.csv', header=None, sep=',')
print(dframe)

# Redirects the file to the standard output and change the separator
# columns allows us to select the specific columns to use
dframe.to_csv(sys.stdout, sep=';', columns=[0,1]) 
```

## Working with JSON

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import json

json_obj="""
{
  "squadName": "Super hero squad",
  "homeTown": "Metro City",
  "formed": 2016,
  "secretBase": "Super tower",
  "active": true,
  "members": [
    {
      "name": "Molecule Man",
      "age": 29,
      "secretIdentity": "Dan Jukes",
      "powers": [
        "Radiation resistance",
        "Turning tiny",
        "Radiation blast"
      ]
    }]
}
"""

data = json.loads(json_obj) # The Json object 'data' is like a dicitionary
print(data['members'][0]['name'])

json.dumps(data) # Method to turn the data back to a string
```
